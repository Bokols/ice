### Preparación de datos

Con base en las observaciones descritas anteriormente, en esta sección voy a hacer las correcciones necesarias para poder realizar mi análisis.

```python
# rename column
df.columns = df.columns.str.lower()
```

```python
# convert year_of_release to integer
df['year_of_release'] = df['year_of_release'].astype('Int64')
```

```python
# remove rows with missing year_of_release and year_of_release < 2000
df = df.loc[(df['year_of_release'] != None)]
```

```python
# change tbd value in user_score to NaN 
df.loc[df['user_score'] == 'tbd', 'user_score'] = None
```

```python
# convert user_score to float64 and scale 100
df['user_score'] = df['user_score'].astype('float64')*10
```

```python
# change outdated esrb rating value
df.loc[df['rating'] == 'EC', 'rating'] = 'E'
df.loc[df['rating'] == 'K-A', 'rating'] = 'M'
df.loc[df['rating'] == 'AO', 'rating'] = 'M'
df.loc[df['rating'] == 'RP', 'rating'] = 'E10+'
```

```python
# create total_sales column
df['total_sales'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales']
```

### El número de juegos lanzados cada año.

```python
plt.figure(figsize=(30,10))
a = sns.countplot(x=df['year_of_release'], palette='flare')

for g in a.patches:
    a.annotate(format(g.get_height(), '.1f'),
                   (g.get_x() + g.get_width() / 2., g.get_height()),
                   ha = 'center', va = 'center', 
                   xytext = (0, 10), rotation=0, 
                   textcoords = 'offset points')
plt.title('¿Cuántos juegos se lanzaron cada año?',fontsize=15)
plt.xlabel('Año')
plt.ylabel('Numero de juegos')
plt.show()
```

#### Observaciones
* A medida que comenzó el nuevo milenio, la cantidad de juegos lanzados se expandió sustancialmente, ya que los juegos se convirtieron en una actividad más popular entre los niños y los jóvenes adolescentes.

* La innovación tecnológica en la nueva década produce una brecha que descontinúa muchas plataformas de juegos, lo que resulta en un patrón decreciente de lanzamientos de juegos.

### Cómo difieren las ventas entre plataformas.

```python
sns.set(context='notebook', style='whitegrid',rc={'figure.figsize':(20,30)})

platform_order_mean = df.groupby('platform')['total_sales'].agg('mean').reset_index().sort_values('total_sales', ascending=False)

b = sns.boxplot(data=df,x='total_sales',y='platform', order=platform_order_mean['platform'])
_ = b.set_title('Distribución de todas las ventas de juegos agrupadas por plataforma')
_ = b.set(xlabel='Ventas globales (millones de USD)', ylabel='plataforma')
```

#### Observaciones
* Distribución de ventas globales
* Si solo se consideran las ventas promedio, Gameboy supera a todas las demás plataformas en ventas globales.
* No obstante, debemos señalar que muchas plataformas, como Wii, la serie Playstation y Nintendo DS, lanzan una cantidad significativa de visitas, lo que aumenta sus ingresos totales.
* El diagrama de distribución a continuación indica que muchos juegos publicados no se venden bien, lo que resulta en un rango bajo de IQR y una gran cantidad de valores atípicos.

### Ventas globales promedio

```python
plt.figure(figsize=(20,5))
c = sns.barplot(data=df,x='platform',y='total_sales', order=platform_order_mean['platform'], ci=None, palette='flare')

for g in c.patches:
    c.annotate(format(g.get_height(), '.1f'),
                   (g.get_x() + g.get_width() / 2., g.get_height()),
                   ha = 'center', va = 'center', 
                   xytext = (0, 10), rotation=0, 
                   textcoords = 'offset points')
plt.title('Ventas promedio en varias plataformas',fontsize=15)
plt.xlabel('plataformas')
plt.ylabel('Ventas globales promedio (millones de USD)')
plt.show()
```

### Ventas globales totales

```python
platform_order_sum = df.groupby('platform')['total_sales'].agg('sum').reset_index().sort_values('total_sales', ascending=False)

plt.figure(figsize=(25,7))
d = sns.barplot(data=df,x='platform',y='total_sales', estimator=sum, order=platform_order_sum['platform'], ci=None, palette='flare')

for g in d.patches:
    d.annotate(format(g.get_height(), '.1f'),
                   (g.get_x() + g.get_width() / 2., g.get_height()),
                   ha = 'center', va = 'center', 
                   xytext = (0, 10), rotation=0, 
                   textcoords = 'offset points')
plt.title('¿Qué plataforma tiene más ventas?',fontsize=15)
plt.xlabel('plataforma')
plt.ylabel('Ventas globales totales (millones de USD)')
plt.show()
```

#### Observaciones

## Evaluación de la hipótesis
Hay dos hipótesis a probar en este proyecto:

* Las calificaciones promedio de los usuarios de las plataformas Xbox One y PC son iguales.
* Las calificaciones promedio de los usuarios de los géneros Acción y Deportes no son iguales.

### Hipótesis 1

* H0 (hipótesis nula): las calificaciones promedio de los usuarios de las plataformas Xbox One y PC son iguales.
* H1 (hipótesis alternativa): las calificaciones promedio de los usuarios de las plataformas Xbox One y PC no son iguales.

El umbral (alfa) se establece en 5\% (2,5\% en cada lado) ya que es una de las medidas comunes de significación estadística.

```python
xone_user_ratings = df[df['platform'] == 'XOne']['user_score'].dropna()
pc_user_ratings = df[df['platform'] == 'PC']['user_score'].dropna()

res_1 = st.levene(xone_user_ratings, pc_user_ratings, center='median', proportiontocut=0.05)
res_1.statistic
```

```python
k, n = 2, 60   # number of samples, total number of observations
dist = st.f(dfn=k-1, dfd=n-k)
val = np.linspace(0, 5, 100)
pdf = dist.pdf(val)

def plot(ax):  # we'll re-use this
    ax.plot(val, pdf, color='C0')
    ax.set_title("Distribución nula de la prueba de Levene")
    ax.set_xlabel("estadística")
    ax.set_ylabel("densidad de probabilidad")
    ax.set_xlim(0, 5)
    ax.set_ylim(0, 1)

fig, ax = plt.subplots(figsize=(8, 5))
pvalue = dist.sf(res_1.statistic)
annotation = (f'p-value={pvalue:.5f}\n(Area sombreada)')
props = dict(facecolor='black', width=1, headwidth=5, headlength=8)
_ = ax.annotate(annotation, (1.5, 0.22), (2.25, 0.3), arrowprops=props)
i = val >= res_1.statistic
ax.fill_between(val[i], y1=0, y2=pdf[i], color='C1')
plot(ax)
plt.show()
```

```python
def statistic(xone_user_ratings, pc_user_ratings):
    return st.levene(xone_user_ratings, pc_user_ratings).statistic
ref = st.permutation_test(
    (xone_user_ratings, pc_user_ratings), statistic,
    permutation_type='independent', alternative='greater'
)
fig, ax = plt.subplots(figsize=(8, 5))
plot(ax)
bins = np.linspace(0, 5, 25)
ax.hist(
    ref.null_distribution, bins=bins, density=True, facecolor="C1"
)
ax.legend(['aproximación asintótica\n(muchas observaciones)',
           'distribución aleatoria nula'])
plot(ax)
plt.show()
```

```python
alpha = 0.05

print('valor_p:',dist.sf(res_1.statistic)) 

if dist.sf(res_1.statistic) < alpha:
    print("Rechazamos la hipótesis nula")
else:
    print("No podemos rechazar la hipótesis nula")
```

#### Conclusión
No podemos rechazar la hipótesis nula en esta prueba, lo que implica que las calificaciones promedio de los usuarios de las plataformas Xbox One y PC no son iguales.
